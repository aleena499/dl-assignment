```python
import pandas as pd
a=pd.read_csv("CarPrice_Assignment.csv",sep=',')
```

```python
a=pd.get_dummies(a)
```

```python
import numpy as np
train=a.iloc[:103]
train_l=train.loc[:,'price']
train_l=np.asarray(train_l)
train_l
```

```python
test=a.iloc[103:165]
test_l=test.loc[:,'price']
test_l=np.asarray(test_l)
test_l
```

```python
val=a.iloc[165:]
val_l=val.loc[:,'price']
val_l=np.asarray(val_l)
val_l
```

```python
train=train.drop(['price'],axis=1)
test=test.drop(['price'],axis=1)
val=val.drop(['price'],axis=1)
```

```python
train
```

```python
val
```

```python
from keras import models,layers
model=models.Sequential()
model.add(layers.Dense(10,activation='relu',input_shape=(200,)))
model.add(layers.Dense(8,activation='relu'))
model.add(layers.Dense(6,activation='tanh'))
model.add(layers.D)
```

```python
model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
```

```python
al=model.fit(train,train_l,
            batch_size=20,
            epochs=100,
            validation_data=(val,val_l))
```
